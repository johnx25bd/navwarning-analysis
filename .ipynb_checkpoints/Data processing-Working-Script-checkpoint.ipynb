```python
# Reading in the message data
with open('C:/Users/Krist/University College London/Spatio-Temporal Data Analysis and Data Mining/WordCount.txt','r') as data:
    text = data.readlines()
    
messageWordCount = {}
for line in text:
    temp = re.split('[\n \t]',line)
    temp = [i for i in temp if i != '']
    try:
        messageWordCount[temp[0]] = int(temp[1])
    except:
        print(line)
```

```python
# Reading in the area data
with open('C:/Users/Krist/University College London/Spatio-Temporal Data Analysis and Data Mining/AreaCount.txt','r') as data:
    text = data.readlines()
    
areaWordsCount = {}
for line in text:
    temp = re.split('[\n \t]',line)
    temp = [i for i in temp if i != '']
    try:
        areaWordsCount[temp[0]] = int(temp[1])
    except:
        print('\n\n',temp)
        print('\n\n',line)
```

```python
areaWordsSubset = {}
areaWordsInMessage = 0
for key in areaWordsCount.keys():
    if key in messageWordCount.keys():
        areaWordsInMessage += 1
        areaWordsSubset[key] = areaWordsCount[key]
#areaWordsSubset
areaWordsInMessage/len(messageWordCount.keys())
```

```python
# Getting initial length
print('The initial length of the wordCount dictionary is: ', len(messageWordCount))
# Remove stopwords from the Message-bag-of-words
# Words we wish to remove:
stopWords
additionalWordsRemove = ['authority','date','cancel','dnc','msg','this','email','phone',
                        "thru","meter","depth","metre","float","top",'fax']
singleletters = [chr(i) for i in range(ord('a'),ord('z')+1)]
months = ['jan','feb','mar','apr','may','jun','jul','aug','sep','oct','nov','dec']
lowerMessageWordCount = {k.lower(): v for k,v in messageWordCount.items() if k.lower() not in stopWords and k.lower() not in additionalWordsRemove and k.lower() not in singleletters and k.lower() not in months}
# We remove all keys that has a length smaller than 3.
lowerMessageWordCount = {k:v for k,v in lowerMessageWordCount.items() if len(k)>2}
# Finally, we remove all words that has a overlap with "Area". Note, we need to adjust for the fact that the keys in "Area" is uppercase letters
lowerMessageWordCount = {k:v for k,v in lowerMessageWordCount.items() if k.upper() not in areaWordsCount.keys()}
```

```python
stop = set(stopwords.words('english'))
exclude = set(string.punctuation) 
lemma = WordNetLemmatizer()

def clean(doc):
    stop_free = " ".join([i for i in doc.lower().split() if i not in stop])
    punc_free = ''.join(ch for ch in stop_free if ch not in exclude)
    normalized = " ".join(lemma.lemmatize(word) for word in punc_free.split())
    noNumeric = [re.sub('[0-9]','',i) for i in normalized.split()]
    abovethree = [i for i in noNumeric if len(i)>2]
    noAddtional = [i for i in abovethree if i not in additionalWordsRemove]
    noSingle = ' '.join([i for i in noAddtional if i not in singleletters])
    
    return noSingle
```

```python
classification = {'Operations':0,'Military':0,'Warnings':0,'Emergency':0}

operations = ['scientific','hazardous']
military = ['gunnery','missile']
warnings = ['iceberg','unlit']
emergency = ['assistance','overboard']
slack = messages.shape[0]*0.9

# operationsWords = {}
# operationsObs = []
# militaryWords = {}
# militaryObs = []
# warningsWords = {}
# warningsObs = []
# emergencyWords = {}
# emergencyObs = []

# The idea is that we start with some "initial beliefs" about the relevant words to classify the messages
# within the four groups defined above. We go through the messages, count how many we classify, and at the end we add new words
# to the individual relevant-words-lists. 

iterations = 0

while(sum(classification.values())<slack):
    
    #### Not sure if it is overkill to initiate the containers every time.
    operationsWords = {}
    operationsObs = []
    militaryWords = {}
    militaryObs = []
    warningsWords = {}
    warningsObs = []
    emergencyWords = {}
    emergencyObs = []
    
    if iterations == 0:
        
        print('\n The initial words in "Operations" are: ', operations)
        print('\n The initial words in "Military" are: ', military)
        print('\n The initial words in "Warnings" are: ', warnings)
        print('\n The initial words in "Emergency" are: ', emergency)    
    
    for obs in messages.index:

        # Checking the occurence of operations-related messages
        if any([True if ele in messages['Message'].loc[obs].lower() else False for ele in operations]):
            classification['Operations'] += 1
            temp = clean(messages['Message'].loc[obs]).split()
            # Word counting based on the operation-related-messages
            for word in temp:
                if word not in operations and word not in months:
                    if word not in operationsWords.keys():
                        operationsWords[word] = 1
                    else:
                        operationsWords[word] += 1

            operationsObs.append(obs)
            
        # Checking the occurence of military-related messages
        if any([True if ele in messages['Message'].loc[obs].lower() else False for ele in military]):
            classification['Military'] += 1
            temp = clean(messages['Message'].loc[obs]).split()
            # Word counting based on the military-related-messages
            for word in temp:
                # To disregard the words that we already use to extract the messages
                if word not in military and word not in months:
                    if word not in militaryWords.keys():
                        militaryWords[word] = 1
                    else:
                        militaryWords[word] += 1
        
            militaryObs.append(obs)
            
        # Checking the occurence of warnings-related messages
        if any([True if ele in messages['Message'].loc[obs].lower() else False for ele in warnings]):
            classification['Warnings'] += 1
            temp = clean(messages['Message'].loc[obs]).split()
            # Word counting based on the warnings-related-messages
            for word in temp:
                # To disregard the words that we already use to extract the messages
                if word not in warnings and word not in months:
                    if word not in warningsWords.keys():
                        warningsWords[word] = 1
                    else:
                        warningsWords[word] += 1
        
            warningsObs.append(obs)
            
        # Checking the occurence of emergency-related messages
        if any([True if ele in messages['Message'].loc[obs].lower() else False for ele in emergency]):
            classification['Emergency'] += 1
            temp = clean(messages['Message'].loc[obs]).split()
            # Word counting based on the emergency-related-messages
            for word in temp:
                # To disregard the words that we already use to extract the messages 
                if word not in emergency and word not in months:
                    if word not in emergencyWords.keys():
                        emergencyWords[word] = 1
                    else:
                        emergencyWords[word] += 1
            
            emergencyObs.append(obs)
            
    sortedOperationWords = sorted(operationsWords.items(), key = lambda kv: kv[1],reverse = True)
    sortedMilitaryWords = sorted(militaryWords.items(), key = lambda kv: kv[1],reverse = True)
    sortedWarningsWords = sorted(warningsWords.items(), key = lambda kv: kv[1],reverse = True)
    sortedEmergencyWords = sorted(emergencyWords.items(), key = lambda kv: kv[1],reverse = True)
    
    # Adding the most frequent word from all the operations-related-messages
    operations.append(sortedOperationWords[0][0])
    #operations.pop(0)
    
    # Adding the most frequent word from all the military-related-messages
    military.append(sortedMilitaryWords[0][0])
    #military.pop(0)
    
    # Adding the most frequent word from all the warnings-related-messages
    warnings.append(sortedWarningsWords[0][0])
    #warnings.pop(0)
    
    # Adding the most frequent word from all the operations-related-messages
    emergency.append(sortedEmergencyWords[0][0])
    #emergency.pop(0)
    
    print('\n The %i iteration resulted in %i observations being classified' % (iterations+1,sum(classification.values())))
    print('\n The words in "Operations" are: ', operations)
    print('\n The words in "Military" are: ', military)
    print('\n The words in "Warnings" are: ', warnings)
    print('\n The words in "Emergency" are: ', emergency)
    iterations += 1
```
